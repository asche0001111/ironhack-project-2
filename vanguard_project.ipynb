```python
#confirm if the difference in completion rate of the new design and the old design is statistically significant
```

```python
completion_rate_ctrl = dfx_final[dfx_final["Variation"] == "Control"]["IsFinal"].sum() / len(dfx_final[dfx_final["Variation"] == "Control"])
```

```python
completion_rate_ctrl
```

```python
completion_rate_test = dfx_final[dfx_final["Variation"] == "Test"]["IsFinal"].sum() / len(dfx_final[dfx_final["Variation"] == "Test"])
```

```python
completion_rate_test
```

```python
#H0 : no statistical difference
#H1 : test statistically varies from control
```

```python
alpha = 0.05
contingency = pd.crosstab(dfx_final["Variation"], dfx_final["IsFinal"])
```

```python
contingency
```

```python
chi2, p_value, dof, expected = chi2_contingency(contingency)

# Interpret the results
print("Chi-square statistic:", chi2)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies:")
print(expected)
```

```python
if p_value < alpha:
    print("Null hypothesis rejected")
else:
    print("Null hypothesis failed to reject")
#proven to be statistically significant due to very low p value, hence the new version is statistically different
```

```python
#conduct a test to see if there is a difference between the finalized completion times of the old and new design

#H0 = there is no statistical difference between the average time of completion of the old and new website design
#H1 = The new design takes less time on average

ctrl_time = dfx_final[(dfx_final["Variation"] == "Control") & (dfx_final["IsFinal"] == True)]["total_process_time_seconds"].mean()
test_time = dfx_final[(dfx_final["Variation"] == "Test") & (dfx_final["IsFinal"] == True)]["total_process_time_seconds"].mean()
time_std = dfx_final[dfx_final["IsFinal"] == True]["total_process_time_seconds"].std()
time_n = dfx_final[dfx_final["IsFinal"] == True]["total_process_time_seconds"].count()
```

```python
print(ctrl_time, test_time, time_std, time_n)
```

```python
dfx_final[dfx_final["IsFinal"] == True]["total_process_time_seconds"].quantile([0.25, 0.5, 0.75])
```

```python
import scipy.stats as st
stat1 = (test_time - ctrl_time)/(time_std/np.sqrt(time_n))
p_value = st.t.sf(abs(stat1), time_n-1)
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Assuming you have imported your DataFrame full_client_profiles

# Subset the DataFrame to include only the columns you need
subset = full_client_profiles[['IsFinal', 'clnt_tenure_yr']]

# Drop any rows with missing values
subset.dropna(inplace=True)

# Separate the two groups for comparison based on the values of 'IsFinal'
group1 = subset[subset['IsFinal'] == True]['clnt_tenure_yr']
group2 = subset[subset['IsFinal'] == False]['clnt_tenure_yr']

# Perform the t-test
t_stat, p_value = ttest_ind(group1, group2)

# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

```
